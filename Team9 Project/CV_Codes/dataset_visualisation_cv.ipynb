```python
import os
import matplotlib.pyplot as plt

# Set the path to your dataset directory
dataset_path = '/content/drive/MyDrive/computer vision/gaussian_filtered_images'

# Get the list of class directories
class_directories = os.listdir(dataset_path)

# Initialize a dictionary to store the count of images per class
class_counts = {}

# Traverse through each class directory
for class_dir in class_directories:
    # Construct the full path to the class directory
    class_path = os.path.join(dataset_path, class_dir)

    # Count the number of images in each class
    num_images = len(os.listdir(class_path))

    # Store the count in the dictionary
    class_counts[class_dir] = num_images

# Plot a pie chart
labels = class_counts.keys()
sizes = class_counts.values()

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

# Set the title of the pie chart
plt.title('Distribution of Images in Each Class')

# Show the pie chart
plt.show()
```
