```python
import os
import cv2
from PIL import Image
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

def apply_gabor_filter(image):
    kernel = cv2.getGaborKernel((21, 21), 8.0, np.pi / 4, 10.0, 0.5, 2, ktype=cv2.CV_32F)
    filtered_image = cv2.filter2D(image, cv2.CV_8UC3, kernel)
    return filtered_image

input_folder = '/content/drive/MyDrive/computer vision/grayscale_images/grayscale_images'

output_folder = '/content/drive/MyDrive/computer vision/gaborfinal'

os.makedirs(output_folder, exist_ok=True)

subfolders = ['Mild', 'Moderate', 'No_DR', 'Proliferate_DR', 'Severe']

for subfolder in subfolders:
    input_subfolder = os.path.join(input_folder, subfolder)
    output_subfolder = os.path.join(output_folder, subfolder)

    os.makedirs(output_subfolder, exist_ok=True)

    for filename in tqdm(os.listdir(input_subfolder), desc=subfolder):
        grayscale_image_path = os.path.join(input_subfolder, filename)
        grayscale_image = Image.open(grayscale_image_path).convert('L')
        grayscale_array = np.array(grayscale_image)

        colored_image = cv2.applyColorMap(grayscale_array, cv2.COLORMAP_VIRIDIS)

        gabor_filtered_image = apply_gabor_filter(colored_image)
        gabor_filtered_image=cv2.cvtColor(colored_image, cv2.COLOR_BGR2RGB)
        output_image_path = os.path.join(output_subfolder, filename)
        Image.fromarray(gabor_filtered_image).save(output_image_path)

```
